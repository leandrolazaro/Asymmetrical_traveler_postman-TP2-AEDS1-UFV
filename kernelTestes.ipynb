```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.patches as mpatches
from matplotlib.legend_handler import HandlerLine2D

plt.style.use('ggplot')
```

```python
registro_tempo=pd.read_csv("tempo.csv")
```

```python
registro_tempo.dtypes
```

```python
registro_tempo.rename(columns=)
```

```python
(registro_tempo.sort_values(by='Tempo (em segundos)').reset_index()['Tempo (em segundos)']).plot(figsize=(15,7))

# Setting the title and labels
plt.title("Tempo de execucao do algoritimo")
plt.xlabel("\nQuantidade de testes feitos")

plt.legend(handler_map={line1: HandlerLine2D(numpoints=4)})

# Ploting the graph
plt.show()
```

```python
(registro_tempo.sort_values(by='Entrada (unidades)').reset_index()['Entrada (unidades)']).plot(figsize=(15,7), color='blue')

# Setting the title and labels
plt.title("Entrada do algorítimo")
plt.xlabel("\nQuantidade de testes feitos")

plt.legend(handler_map={line1: HandlerLine2D(numpoints=4)})

# Ploting the graph
plt.show()
```
